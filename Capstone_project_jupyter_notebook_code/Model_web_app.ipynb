```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("mushrooms.csv")
```

```python
data = data.sample(frac=1)
```

```python
X = data.drop(["class"], axis =1)

y = data["class"]
```

```python
for attr in X.columns:
    print("\n*", attr, "*")
    print(X[attr].value_counts())
```

```python
X.drop(["veil-type"], axis=1, inplace = True)
```

```python
for col in X.columns:
    if len(X[col].value_counts()) ==2:
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col])
```

```python
X.head()
```

```python
X = X[["bruises", 'gill-attachment', 'gill-spacing', 'gill-size', 'stalk-shape']]
X.reset_index(inplace = True)
X
```

```python
X.drop("index", axis = True, inplace= True)
```

```python
#X = pd.get_dummies(X)
```

```python
X.head(2)
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =.2)
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
logreg_pred = logreg.predict(X_test)
```

```python
logreg.score(X_test, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C = 1.0, penalty = "l2")
lr.fit(X_train,y_train)
print("Logistic Regression Max Score : ",lr.score(X_test,y_test))
lr_max = lr.score(X_test,y_test)
```

```python
from joblib import dump, load

dump(logreg, filename="Capstone_mushroom.joblib")
```
